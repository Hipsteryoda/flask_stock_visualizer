```python
# Assemble their history
built_df = analysis.build_stocks_df(df)
built_df
```

```python
# Get moving average and bolinger bands
ma_df = analysis.n_day_moving_average(built_df, 20)
bol_df = analysis.bolinger_bands(ma_df, '20_day_moving_average', 20)
```

```python
plot = analysis.plotly_plot_bolinger(bol_df, bol_df['ticker'][0], 20)
type(plot)
```

```python
plot.show()
```

```python
from plotly.subplots import make_subplots
import plotly.graph_objects as go
x_axis=bol_df[bol_df['ticker']==bol_df['ticker'][0]].index
fig = make_subplots(specs=[[{"secondary_y": True}]])
fig.add_trace(
    go.Scatter(x=x_axis, y=bol_df[bol_df['ticker']==bol_df['ticker'][0]]['Close'], name='Close Price')
    )
type(fig)
```

```python
import json
data = [fig]
graph = json.dumps(data, cls=plotly.utils.PlotlyJSONEncoder)
type(graph)
```
